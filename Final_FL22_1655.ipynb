This information can be summarized concisely in a [confusion matrix](https://www.w3schools.com/python/python_ml_confusion_matrix.asp)  (we saw this in class24).
```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler  
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay

skyserv_new_r = pd.read_csv("skyserver.csv")
skyserv_new_r.drop(['u', 'g', 'i','z','redshift'], axis=1, inplace=True)
#skyserv_drop_GALX = skyserv_new.loc[skyserv_new_r["class"] != 'GALAXY']


X_starserv = skyserv_new_r.drop("class", axis=1) 
X_starserv.shape

y_starserv =  skyserv_new_r["class"]
y_starserv.shape


stX_train, stX_test, styy_train, styy_test = train_test_split(X_starserv , y_starserv, random_state=1)


scaler = StandardScaler()  


scaler.fit(stX_train)  

Xtrain = scaler.transform(stX_train)  

Xtest = scaler.transform(stX_test) 

model_starserv = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(60, 30), random_state=1, max_iter=1000)

model_starserv.fit(stX_train, styy_train)

styy_model = model_starserv.predict(stX_test)


print('\n STAR or QSO serv confusion matrix:\n',confusion_matrix(styy_test, styy_model),'\n\n')

print('classification_report\n',classification_report(styy_test, styy_model),'\n\n')

round(accuracy_score(styy_test, styy_model),3)
```

There is about a 60.2% accuaracy for the impurity percentage which is relativley low. The false positives could be due to many stars and les QSOs. Pulled the classification report and confusion matrix by utilizing learned code from class24.

## i) Training the Neural Network
Train a neural network to classify stars and galaxies using the reduced dataset that you made in part h. The question is, how well can you classify star v/s galaxy based on the 5 bands of light data from the SDSS? Play with the hyperparameters a little and see how it affects the accuracy_score to make sure you have a somewhat optimized network. Remember to scale data, since this is crucial in the MLP (class24). **Show the farction of correctly classified samples.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler  
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay


X_skyserv = skyserv_drop.drop("class", axis=1)
X_skyserv.shape

#Target vector
y_skyserv = skyserv_drop["class"]
y_skyserv.shape

skyX_train, skyX_test, skyy_train, skyy_test = train_test_split(X_skyserv, y_skyserv, random_state=1)


scaler = StandardScaler()  

scaler.fit(skyX_train)  

Xtrain = scaler.transform(skyX_train)  

Xtest = scaler.transform(skyX_test)  

model_skyserv = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(60, 30), random_state=1, max_iter=1000)

model_skyserv.fit(skyX_train, skyy_train)

skyy_model = model_skyserv.predict(skyX_test)

round(accuracy_score(skyy_test, skyy_model),3)

print('\n skyserv confusion matrix:\n',confusion_matrix(skyy_test, skyy_model),'\n\n')



skyserv_cm=confusion_matrix(skyy_test, skyy_model)
disp = ConfusionMatrixDisplay(confusion_matrix=skyserv_cm, display_labels=model_skyserv.classes_)
disp.plot()
plt.show()

round(accuracy_score(skyy_test, skyy_model),3)
```
